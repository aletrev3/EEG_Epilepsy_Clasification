```python
import matplotlib.pyplot as plt
#Comparison betwwen a non-seizure record and a seizure record (2 seconds each)
#Non-seizure file chb01_14
file_14 = mne.io.read_raw_edf('/content/drive/MyDrive/datos_eeg/chb01/chb01_14.edf')
data_14 = file_14.get_data()
#plotting file
chb01_14 = plt.plot(data_14[0, 445440:445952])
plt.title("A. Seconds 1740 to 1742 of chb01_14 as raw data (electrode 0)")
plt.show()

#Non-seizure file chb01_15
file_15 = mne.io.read_raw_edf('/content/drive/MyDrive/datos_eeg/chb01/chb01_15.edf')
data_15 = file_15.get_data()
#plotting file
chb01_15 = plt.plot(data_15[0, 445440:445952])
plt.title("A. Seconds 1740 to 1742 of chb01_15 as raw data (electrode 0)")
plt.show()
```
